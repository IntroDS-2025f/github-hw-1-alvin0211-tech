````python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(768)
n = 200


x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # Âü∫Á§éÈªûÈù¢Á©ç

size_scale = 5
sizes = size_var * size_scale


scatter = plt.scatter(x, y, c=y, s=sizes, cmap='viridis', alpha=0.7, edgecolor='k')


plt.title("Scatter with Colormap")
plt.xlabel("x")
plt.ylabel("y")


cbar = plt.colorbar(scatter)
cbar.set_label("y-values")

plt.show()

````
